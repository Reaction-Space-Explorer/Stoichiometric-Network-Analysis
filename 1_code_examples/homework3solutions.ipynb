where \\(K_d = k_{-1}/k_1\\). __Hint:__ Take into account the mass conservation and the fact that the quasi-equilibrioum state is valid when \\(ds_0/dt = 0\\)

> 1. Dynamic Mass Balances (3 pts)
$$\begin{align}
\frac{ds}{dt} &= -k_1es + k_{-1}x \\ 
\frac{dx}{dt} &= k_1es - k_{-1}x - k_2x \\
\frac{dp}{dt} &= k_2x \end{align}$$
> 2. Time invariants (2 pts)
$$\begin{align}e_0 = e + x &\Leftrightarrow e = e_0 - x \\

<font color='red'> **(i)** Solution (6 pts) </font> 

> Apply QSSA (i.e. \\(dx/dt = 0\\)) (1 pt)
$$\begin{align}\frac{dx}{dt} = 0 
&= k_1es - k_{-1}x - k_2x, \ \ \ \text{Substitute enzyme conservation:} e = e_0 - x  \\ 
&= k_1(e_0 - x)s - k_{-1}x - k_2x \\ &= k_1e_0s - k_1xs - k_{-1}x - k_2x \\ 
&= k_1(e_0 - x)s - k_{-1}x - k_2x \\ &= k_1e_0s - x(k_1s + k_{-1} + k_2) \Leftrightarrow x = \frac{k_1e_0s}{(k_1s + k_{-1} + k_2)} (2 pts)
\end{align}$$
> Plug solution for \\(x\\) into rate of change for substrate: 
$$\begin{align} \frac{ds}{dt} 
&= -k_1es + k_{-1}x \\ 
&= -k_1(e_0 - x)s + k_{-1}x \\ 
&= -k_1e_0s + k_1xs + k_{-1}x \\ 
&= -k_1e_0s + x(k_1s + k_{-1}) \\ 
&= -k_1e_0s + \frac{(k_1s + k_{-1})k_1e_0s}{(k_1s + k_{-1} + k_2)} \\
&= \frac{-k_1e_0s(k_1s + k_{-1} + k_2) + (k_1s + k_{-1})k_1e_0s}{(k_1s + k_{-1} + k_2)} \\
&= \frac{-k_2k_1e_0s}{(k_1s + k_{-1} + k_2)},\ \ \ \text{Multiply by }\frac{1/k_1}{1/k_1} \\
&= \frac{k_2e_0s}{s + \frac{k_{-1} + k_2}{k_1}}, \ \ \ \text{Define Michaelis-Menten Constant: }K_m = \frac{k_{-1} + k_2}{k_1} (1 pt) \\ 
\frac{ds}{dt} &= \frac{-k_2e_0s}{K_m + s} (2 pts)
\end{align}$$

<font color='red'> **(ii)** Solution (6 pts) </font> 

> Apply QEA (i.e. \\(ds_0/dt = 0\\)) (1 pt)
$$\begin{align}
\frac{ds_0}{dt} = 0 &= -k_1es + k_{-1}x, \ \ \ \text{Substitute enzyme conservation:} e = e_0 - x  \\ 
&= -k_1(e_0 - x)s + k_{-1}x \\ 
&= -k_1e_0s - k_1xs + k_{-1}x \\ 
0 &= -k_1e_0s - x(k_1s + k_{-1})\ \Leftrightarrow x = \frac{k_1e_0s}{k_1s + k_{-1}}
\end{align} \\ 
\text{Define Michaelis-Menten Constant: }K_d = \frac{k_{-1}}{k_1} (1 pt) \\ 
x = \frac{e_0s}{K_d + s} (2 pts)
$$

> From the substrate conservation:
$$\begin{align}s &= s_0 - x - p \\ 
\frac{d}{dt}(s &= s_0 - x - p)\\ \\
\frac{ds}{dt} &= \frac{ds_0}{dt} - \frac{dx}{dt} - \frac{dp}{dt} \\
&= 0 - \frac{d}{dt}(x) - k_2x \\
&= 0 - \frac{d}{dt}(\frac{e_0s}{K_d + s}) - k_2(\frac{e_0s}{K_d + s}) \\
\end{align}
$$

> Note that the derivative
$$\frac{d}{dt}(\frac{s}{K_d + s}) = (\frac{1}{K_d + s} - s\frac{1}{(K_d + s)^2})\frac{ds}{dt}
$$

> Therefore
$$\begin{align}
\frac{ds}{dt} &= - e_0\frac{d}{dt}(\frac{s}{K_d + s}) - k_2(\frac{e_0s}{K_d + s}) \\
\frac{ds}{dt} &= - \frac{ds}{dt}e_0(\frac{1}{K_d + s} - s\frac{1}{(K_d + s)^2}) - k_2(\frac{e_0s}{K_d + s}) \\
\frac{ds}{dt} + \frac{ds}{dt}e_0(\frac{1}{K_d + s} - s\frac{1}{(K_d + s)^2}) &= k_2(\frac{e_0s}{K_d + s}) \\
\frac{ds}{dt}(1 + e_0(\frac{1}{K_d + s} - s\frac{1}{(K_d + s)^2})) &= k_2(\frac{e_0s}{K_d + s}) \\
\frac{ds}{dt} &= \frac{-k_2(\frac{e_0s}{K_d + s})}{1 + e_0(\frac{1}{K_d + s} - s\frac{1}{(K_d + s)^2})} \\
\frac{ds}{dt} &= \frac{-k_2(\frac{e_0s}{K_d + s})}{1 + \frac{e_0K_d}{(K_d + s)^2}} \\
\frac{ds}{dt} &= \frac{-k_2e_0s}{K_d + s + \frac{e_0K_d}{K_d + s}} (2 pts)
\end{align}$$

```python
model = MassModel('MM_Open')

## Define metabolites
s = MassMetabolite("s")
e = MassMetabolite("e")
x = MassMetabolite("x")
p = MassMetabolite("p")

# Define reactions
b1 = MassReaction("b1", reversible = False)
v1 = MassReaction("v1")
v2 = MassReaction("v2", reversible = False)
v3 = MassReaction("v3", reversible = False)

# Add metabolites to reaction
b1.add_metabolites({s: 1})
v1.add_metabolites({s: -1, e: -1, x: 1})
v2.add_metabolites({x: -1, e: 1, p: 1})
v3.add_metabolites({p: -1})
model.add_reactions([b1, v1, v2, v3])

## Define parameters
b1.kf = 1
v1.kf, v1.kr = (1000, 1000)
v2.kf = 100
v3.kf = 10

# Set initial conditions for model
s.ic = 10
e.ic = 10
x.ic = 0
p.ic = 0

# Utilize type 2 rate law for kf and kr parameters defined
model.get_rate_expressions(rate_type = 2, update_reactions = True)

# Set a custom rate for b1 to remove substrate concentration dependence
model.add_custom_rate(reaction = b1, custom_rate = b1.kf_str)
```
